# Data Analysis and Visualization

```python
cancelled_prec = df['is_canceled'].value_counts()
cancelled_prec
```

```python
#Percentage
cancelled_prec = df['is_canceled'].value_counts(normalize=True)
cancelled_prec
```

```python
#Check cancelled and non cancelled staus

plt.figure(figsize=(5,4))
plt.title('Reservation Status')
plt.bar(['Cancelled','Not Cancelled'],df['is_canceled'].value_counts(), edgecolor='k', width=0.7)
plt.show()
```

```python
#Check cancellations and non cancellations for Resort and city hotel 

plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')
legend_labels,_=ax1.get_legend_handles_labels()
plt.title('Reservation status in different hotels', fontsize=20)
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')
plt.show()
```

```python
#28% cancellation are from Resort Hotel

resort_hotel = df[df['hotel'] =='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)  
```

```python
#42% cancellation are from City Hotel

city_hotel = df[df['hotel'] =='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True) 
```

```python
#Grouping dates with adr 
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
resort_hotel
```

```python
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel
```

```python
# our Assumption was City hotel prices will be low than Resort hotels - will visualize it

plt.figure(figsize = (20,8))
plt.title('Average Daily rate in City and Resort hotel', fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City hotel')
plt.legend(fontsize=10)
plt.show()

#After Visualizing it we got to know that city hotel prices are low compared to resort hotel -- In weekends prices are higher
```

```python
#check in which month we got more cancellations

df['Month']=df['reservation_status_date'].dt.month
```

```python
plt.figure(figsize=(8,5))
plt.title('Reservation status per Month', fontsize=30)
sns.countplot(x='Month', hue='is_canceled', data=df)
plt.legend(['Not cancelled','Cancelled'])
plt.show()

#In january we have more cancellations and in Aug - we have less cancellations
```

```python
month_adr=df[df['is_canceled'] == 1].groupby('Month')['adr'].sum().reset_index()
month_adr
```

```python
plt.figure(figsize=(15,8))
plt.title('ADR for Each Month', fontsize=30)
sns.barplot('Month', 'adr', data=month_adr)
plt.show()

#in August ADR is less - it means If lower prices then less cancellations, if higher prices more cancellations
```

```python
#check top 10 countries which has higher cancellations
cancelled_data = df[df['is_canceled']==1]

noncancelled_data=df[df['is_canceled']==0]

top_10_countries=cancelled_data['country'].value_counts()[:10]
top_10_countries
```

```python
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with reservation cancelled')
plt.pie(x=top_10_countries, labels=top_10_countries.index, autopct='%.2f')
plt.show()

#Portgul country has highest cancellation rates - So we can suggest that country to lower prices,good services, advertisements
```

```python
#Check from where clients are coming?

df['market_segment'].value_counts()

#56K clients are coming from online Travel Agent
```

```python
df['market_segment'].value_counts(normalize=True)
```

```python
#Check ADR for Cancellation and non-cancellation

cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)


noncancelled_df_adr=noncancelled_data.groupby('reservation_status_date')[['adr']].mean()
noncancelled_df_adr.reset_index(inplace=True)
noncancelled_df_adr.sort_values('reservation_status_date',inplace=True)


plt.figure(figsize=(20,8))
plt.title('Average Daily Rate', fontsize=30)
plt.plot(noncancelled_df_adr['reservation_status_date'],noncancelled_df_adr['adr'], label='Non_canceled')

plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label='canceled')
plt.legend()
plt.show()
```

```python
#Take data between 2016 and 2017

cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]

noncancelled_df_adr = noncancelled_df_adr[(noncancelled_df_adr['reservation_status_date']>'2016') & (noncancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
#Visulize ADR for cancellation and non-cancellation between 2016 and 2017

plt.figure(figsize=(20,8))
plt.title('Average Daily Rate', fontsize=30)

plt.plot(noncancelled_df_adr['reservation_status_date'],noncancelled_df_adr['adr'], label='Non_canceled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label='canceled')

plt.legend()
plt.show()
```
